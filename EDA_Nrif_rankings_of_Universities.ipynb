> **CORRELATION**





> **correlation can be calculated only on numerical columns , we can't calculate correlation on non-numeric.**



```python
numeric_features = data.select_dtypes(include = [np.number])
numeric_features.columns
```

**Here numeric columns are stored in variable called number if features and columns are extracted.We can see that nearly 8 columns outof 12 have numerical values.**

They are1)Score 2)TLR 3)RPC 4)GO 5)OI 6)PERCEPTION 7)Year 8)Year_rank

```python
numeric_features.shape, numeric_features1.shape
```

**It gave rows and columns .And by this we can conclude that 8columns are numeric outof 12**

```python
categorical_features = data.select_dtypes(include = [np.object])
categorical_features.columns
```

**It gives all the string columns.We have nearly 4 outof 12columns**

They are1)InstituteID 2)Name 3)City 4)State

```python
correlation = numeric_features.corr()
print(correlation['Year_rank'].sort_values(ascending = False), '\n')
```

**To find the correlation, we are using corr method.**

```python
f, ax = plt.subplots(figsize = (14, 12))
plt.title('Correlation of numeric features with Year_rank')
sns.heatmap(correlation, square=True, vmax=0.8)
```

**This is the correlation matrix for all the 9numerical columns**

```python
k=5
cols = correlation.nlargest(k, 'Year_rank')['Year_rank'].index
print(cols)
```
